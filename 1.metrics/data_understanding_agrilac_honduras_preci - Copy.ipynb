# missing value plots over time for each station

```python
precipitacion_g = precipitacion.copy()
```

```python
for i in precipitacion.columns:
    j=precipitacion_g.columns.get_loc(f"{i}")+1
    precipitacion_g[f"{i}"] = precipitacion_g[f"{i}"].replace(precipitacion_g[f"{i}"].dropna().unique(), j)
```

```python
# Creamos la gráfica
# Creamos la figura y el eje

fig, ax = plt.subplots(figsize=(10, 10))
for col in precipitacion_g.columns:
    fig = plt.plot(precipitacion_g.index, precipitacion_g[col], marker='o', label=col)

ax.set_yticks(list(range(1, 48)))
ax.set_yticklabels(precipitacion_g.columns.to_list());

#fig.show()
```
