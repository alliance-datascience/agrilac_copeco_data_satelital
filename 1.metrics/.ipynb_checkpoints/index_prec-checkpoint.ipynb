# Outputs

```python
def count_days_greater_than_limit(x, limit):
    if x.dropna().size / x.size >= 0.95:
        return (x > limit).sum()
    else:
        return np.nan
```

```python
grouper = pd.Grouper(freq='M')
```

```python
precipitacion.groupby(grouper).agg(lambda x: count_days_greater_than_limit(x, 10)).to_csv('Z:\\1.Data\\Honduras\\transformed\\monthly_agg\\obs_stations_r10mm_monthly.csv')
```

```python
precipitacion.groupby(grouper).agg(lambda x: count_days_greater_than_limit(x, 15)).to_csv('Z:\\1.Data\\Honduras\\transformed\\monthly_agg\\obs_stations_r15mm_monthly.csv')
```
